```python
#lets make a plan to clean the 'cateogry and subcateogry' column

#we can split the values in this column and add the values to new columns: crowdfunding_info_df['cateogry'] and crowdfunding_info_df['subcategory']

# we're gonna make one split on the forward slash line



```

```python
# Assign the category and subcategory values to category and subcategory columns.
crowdfunding_info_df[["category","subcategory"]] = crowdfunding_info_df["category & sub-category"].str.split('/', n=1, expand=True)
crowdfunding_info_df.head(10)

# in the split function, the number of splits is n and its set equal to 1 (n=1)

#the expand = True parameter is required to finalize the column split and send the data to the new columns
```

```python
# we want to create a unique identifier for each category and subcategory column

# to do this, you need to use the .nunique() function on a dataframe column.
# nunique function will return the number of unique values in that column

# Get the distinct values in the category and subcategory columns.
print(crowdfunding_info_df["category"].nunique())
print(crowdfunding_info_df["subcategory"].nunique())

# we can see the unique values in the category column is 9
# the unique values in the subcategory column is 24

```

```python
# Now create the category and subcategory dataframes

# Get the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df["category"].unique()
subcategories = crowdfunding_info_df["subcategory"].unique()
print(categories)
print()
print(subcategories)

# this will make lists of all the categories and subcategories, every element in each list will become the column header name and column of a new dataframe

```
